```python
# Step 1: Install required libraries
!pip install transformers
!pip install datasets
!pip install torch

# Step 2: Import libraries
from transformers import GPT2Tokenizer, GPT2LMHeadModel, TextDataset, DataCollatorForLanguageModeling, Trainer, TrainingArguments

# Step 3: Upload dataset file
from google.colab import files
uploaded = files.upload()  # Upload 'tech_dataset.txt'

# Step 4: Load tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

# Step 5: Prepare dataset
def load_dataset(file_path, tokenizer):
    return TextDataset(
        tokenizer=tokenizer,
        file_path=file_path,
        block_size=128
    )

def load_data_collator(tokenizer):
    return DataCollatorForLanguageModeling(
        tokenizer=tokenizer, mlm=False
    )

train_dataset = load_dataset("tech_dataset.txt", tokenizer)
data_collator = load_data_collator(tokenizer)

# Step 6: Training arguments
training_args = TrainingArguments(
    output_dir="./gpt2-finetuned-tech",
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=2,
    save_steps=500,
    save_total_limit=2,
)

# Step 7: Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=train_dataset,
)

# Step 8: Train the model
trainer.train()

# Step 9: Save model
trainer.save_model("./gpt2-finetuned-tech")
tokenizer.save_pretrained("./gpt2-finetuned-tech")

# Step 10: Generate new text
from transformers import pipeline
generator = pipeline("text-generation", model="./gpt2-finetuned-tech", tokenizer="./gpt2-finetuned-tech")

print("Generated Text:")
print(generator("Artificial Intelligence", max_length=50, num_return_sequences=1)[0]['generated_text'])
```

```python
def ai_recommend(prompt):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs, max_length=120)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Example usage
print(" Prompt: Suggest a creative AI-based project idea for engineering students")
print(" AI Response:", ai_recommend("Suggest a creative AI-based project idea for engineering students"))

print("\n Prompt: Explain how AI can help in healthcare")
print(" AI Response:", ai_recommend("Explain how AI can help in healthcare"))
```
